```python
new_cols_dict={'timestamp': 'time',
               'cnt':'count',
               't1': 'temp_real_C',
               't2': 'temp_feels_like_C',
               'hum':'humidity_percent',
             'wind_speed':'wind_speed_kph',
            'weather_code':'weather',
            'is_holiday': 'is_holiday',
            'is_weekend':'is_weekend',
                'season':'season'}
```

```python
bikes.columns #looking at all column names inorder to create a dictionary
```

```python
bikes.rename(new_cols_dict,axis=1,inplace=True)
```

```python
bikes
```

```python
bikes.humidity_percent=bikes.humidity_percent/100 #changing the humidity_percent into numbers ranging from 0 to 1
```

```python
bikes
```

```python
season_dict={'0.0':'spring',
             '1.0':'summer',
             '2.0':'autumn',
             '3.0':'winter'
    
}    #creating a dictionary to map the seasons from integers to actual written values
```

```python
weather_dict={'1.0':'Clear',
'2.0':'Scattered clouds',
'3.0':'Broken clouds',
'4.0':'Cloud',
'7.0':'Rain',
'10.0':'Rain with thunderstorms',
'26.0':'Snowfall'} 
   #creating a dictionary to map the wearher from integers to actual written values
```

```python
#changing the season column datatype to a string
bikes.season=bikes.season.astype('str')
#mapping the values 0-3 to the actual written season
bikes.season=bikes.season.map(season_dict)
#changing the weather column datatype to a string
bikes.weather=bikes.weather.astype('str')
#mapping the values to the actual written weather
bikes.weather=bikes.weather.map(weather_dict)
```
